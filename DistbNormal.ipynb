# Codigo python para el calculo de probabilidades normales

Ejemplo básico en Python para trabajar con probabilidades normales usando la librería scipy.stats y matplotlib para graficar:

## Ejemplos: Distribución Normal en Python

```python
# Importación de librerias o bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parámetros de la distribución normal
media = 0      # media (mu)
desv = 1        # desviación estándar (sigma)
```

```python
# Crear un rango de valores x
x = np.linspace(-4, 4, 1000)
```

```python
# Obtener la función de densidad de probabilidad (PDF)
pdf = norm.pdf(x, loc=media, scale=desv)
```

```python
# Graficar la curva normal
plt.plot(x, pdf, label='N(0,1)')
plt.title('Distribución Normal Estándar')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.legend()
plt.show()
```

### Cálculo de la densidad

```python
# Obtener la función de densidad de probabilidad (PDF)
x= 1 # Ingresar el valor de z
pdf = norm.pdf(x, loc=media, scale=desv)
print(pdf)
```

## Calcular probabilidades

### 1. P(X ≤ x)

```python
# P(X ≤ 1.5)
x=3 # Ingresar valor de x
prob = norm.cdf(x, loc=media, scale=desv)
print("P(X ≤ 1.5) =", prob)
```

### 2. P(X > x)

```python
# P(X > 1.5)
x=3 # Ingresar valor de x
prob = 1 - norm.cdf(x, loc=media, scale=desv)
print("P(X > x) =", prob)
```

### P(a < X < b)

```python
# P(a < X < b)
a= -1 #Ingresar el valor de "a"
b= 2 #Ingresar el valor de "b"
prob = norm.cdf(b, loc=media, scale=desv) - norm.cdf(a, loc=media, scale=desv)
print("P(a < X < b) =", prob)
```

## Código: Graficar y Sombrear Probabilidades Normales


### 1. P(X ≤ x)

```python
# Parámetros de la distribución normal
media = 0      # media (mu)
desv = 1        # desviación estándar (sigma)
# Rango de valores en el eje x
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=media, scale=desv)

# Rango a sombrear (por ejemplo: P(X < 1.5)) # Se puede cambiar
x_sombreado = np.linspace(-4, 1.5, 500)
y_sombreado = norm.pdf(x_sombreado, loc=media, scale=desv)

# Crear la gráfica
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Distribución N(0,1)', color='blue')
plt.fill_between(x_sombreado, y_sombreado, color='skyblue', alpha=0.5, label='Área: P(X < 1.5)')

# Etiquetas
plt.title('Distribución Normal con Área Sombreada')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.legend()
plt.show()

# Mostrar la probabilidad numérica
probabilidad = norm.cdf(1.5, loc=media, scale=desv)
print(f"P(X < 1.5) = {probabilidad:.4f}")
```

### 1. P(a<X<b)

```python
# Parámetros de la distribución normal
media = 0      # media (mu)
desv = 1        # desviación estándar (sigma)

# Rango de valores en el eje x
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=media, scale=desv)

# Rango a sombrear (por ejemplo: P(-1 < X < 1.5)) # Se puede cambiar
x_sombreado = np.linspace(-1, 1.5, 500)
y_sombreado = norm.pdf(x_sombreado, loc=media, scale=desv)

# Crear la gráfica
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Distribución N(0,1)', color='blue')
plt.fill_between(x_sombreado, y_sombreado, color='skyblue', alpha=0.5, label='Área: P(-1 < X < 1.5)')

# Etiquetas
plt.title('Distribución Normal con Área Sombreada')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.legend()
plt.show()

# Mostrar la probabilidad numérica
probabilidad = norm.cdf(1.5, loc=media, scale=desv) - norm.cdf(-1, loc=media, scale=desv)
print(f"P(-1 < X < 1.5) = {probabilidad:.4f}")
```

### P(X > x)

```python
# Parámetros de la distribución normal
media = 0      # media (mu)
desv = 1        # desviación estándar (sigma)

# Rango de valores en el eje x
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=media, scale=desv)

# Rango a sombrear (por ejemplo: P(X > 1.5)) # Se puede cambiar
x_sombreado = np.linspace(1.5, 4, 500)
y_sombreado = norm.pdf(x_sombreado, loc=media, scale=desv)

# Crear la gráfica
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Distribución N(0,1)', color='blue')
plt.fill_between(x_sombreado, y_sombreado, color='skyblue', alpha=0.5, label='Área: P(X > 1.5)')

# Etiquetas
plt.title('Distribución Normal con Área Sombreada')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.legend()
plt.show()

# Mostrar la probabilidad numérica
probabilidad = 1-norm.cdf(1.5, loc=media, scale=desv)
print(f"P(-1 < X < 1.5) = {probabilidad:.4f}")
```
